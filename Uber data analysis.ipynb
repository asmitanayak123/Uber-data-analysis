```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv ('uber.csv')
df
```

```python
df.info()
```

```python
#1.In which Unnamed: 0 the last digit is 4 ?

import numpy as np
import pandas as pd
df['Unnamed: 0'] = df['Unnamed: 0'].astype(str)
result = df[df['Unnamed: 0'].str.endswith('4')]
result
```

```python
#2  In which Unnamed: 0 position the pickup_longitude is heighest ?
row_with_highest_longitude = df['pickup_longitude'].idxmax()
print(row_with_highest_longitude)
position_with_highest_longitude = df.at[row_with_highest_longitude, 'Unnamed: 0']

position_with_highest_longitude
```

```python
#3.In fair_amount 7 what are the top 5 passenger_count ?

fare_amount = df[(df['fare_amount']) == 7]
print(fare_amount[['fare_amount','passenger_count']].sort_values(['passenger_count'],ascending = False).head(5))
```

```python
#4.what are the top 5 passenger count ?

top_5_passenger_count = df.sort_values(['passenger_count'],ascending = False).head(5)
top_5_passenger_count
```

```python
#5.What are the key where the pickup_latitude and dropoff_latitude values are not same ?

not_same_values = df[df['pickup_latitude'] != df['dropoff_latitude']]
print(not_same_values[['key','pickup_latitude','dropoff_latitude']])
```

```python
#6.Which pickup_datetime the fare_amount is greater than 7.5 and less than 27.0 ?

fare_amount = df[(df['fare_amount'] > 7.5) & (df['fare_amount'] < 27)]
print(fare_amount[['pickup_datetime','fare_amount']])
```

```python
#7.What are the key where the pickup_longitude and dropoff_longitude values are same ?

same_values = df[df['pickup_longitude'] == df['dropoff_longitude']]
print(same_values[['key','pickup_longitude','dropoff_longitude']])
```

```python
#8.What are the top ten lowest pickup_latitude ?

lowest_top_10 = df.sort_values(['pickup_latitude']).head(10)
print(lowest_top_10['pickup_latitude'])
```

```python
#9.What is the average fare_amount where passenger_count is 1 ?

passenger_count = df[df['passenger_count'] == 1]
print(passenger_count[['fare_amount','passenger_count']])
average_fare_amount = passenger_count['fare_amount'].mean()
print(average_fare_amount)
```

```python
#10.What are the top ten lowest pickup_latitude in fare_amount bellow 6.0 ?

fare_amount_6 = df[df['fare_amount'] < 6]
print(fare_amount_6[['fare_amount','pickup_latitude']].sort_values(['pickup_latitude']).head(10))
```

```python
#11.What are the passenger_count where Unnamed: 0 's last digit is 3 ?

df['Unnamed: 0'] = df['Unnamed: 0'].astype(str)
last_digit_4 = df[df['Unnamed: 0'].str.endswith('3')]
print(last_digit_4[['Unnamed: 0','passenger_count']])
```

```python
#13.Concate the two column by '+' pickup_longitude and pickup_latitude ?

df['pickup_longitude-pickup_latitude'] = df['pickup_longitude'].astype(str) + '+' + df['pickup_latitude'].astype(str)
print(df)
```

```python
#14.Plot the line graph pickup_longitude over pickup_latitude by using the first 10 values ?

First_10values = df[['pickup_longitude','pickup_latitude']].head(10)
print(First_10values)

#Total recovered over time
plt.figure(figsize=(10,8))
plt.plot(First_10values['pickup_longitude'], First_10values['pickup_latitude'], marker='o', linestyle='-')
plt.title('Pickup Longitude over Pickup Latitude')
plt.xlabel('Pickup Longitude')
plt.ylabel('Pickup Latitude')
plt.grid(True)
plt.show()
```

```python
#15 plot a bar graph fare_amount over passenger_count and using last 10 datas ?

last_10_datas = df[['fare_amount','passenger_count']].tail(15)
print(last_10_datas)
plt.figure(figsize=(10, 8))
plt.bar(last_10_datas['fare_amount'],last_10_datas['passenger_count'])
plt.title('bar graph fare_amount over passenger_count')
plt.xlabel('fare_amount')
plt.ylabel('passenger_count')
plt.xticks(rotation=75)
plt.show()
```

```python
#16. Split the two column by date column and time column ?

new = df['key'].str.split()
print(new)
df[['Date','Time']] = pd.DataFrame(new.tolist(), index=df.index)
df['Date'] = new.str[0]
df['Time'] = new.str[1]
df.drop(columns = ['key'],inplace = True)
print(df)
```

```python
# 17. Split the Date column in to year column and date column ?


new = df['Date'].str.split('-',n=1,expand = True)
df[['Year','Date']] = new
print(df)
```

```python
#18.What are the top 5 passenger_count in which time ?


required = df[['passenger_count','Time']]
print(required)
top_five = required.sort_values('passenger_count',ascending = False).head(5)
print(top_five[['passenger_count','Time']])
```

```python
#19.What are the top 5 dates that are used in maximum times ?

used_in_maximum = df['Date'].value_counts()
print(used_in_maximum.head(5))
```

```python
#20.In the year 2009 what is the heighest fare amount ?


in_year_2009 = df[df['Year'] == '2009']
print(in_year_2009[['fare_amount','Year']])
heighest_fare_amount = in_year_2009.sort_values('fare_amount',ascending = False).head(1)
print(heighest_fare_amount[['fare_amount','Year']])
```
